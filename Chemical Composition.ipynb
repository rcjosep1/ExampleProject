# Weight % Calculation from Mol Fraction of Minerals

```python
import numpy as np
```

```python
import periodictable as pt
```

Olivine; Mg2SiO4

Pyroxene; MgSiO3

Mg2SiO4 -> 2MgO + SiO2

MgSiO3 -> MgO + SiO2

```python
pt.formula('MgSiO4').mass
```

```python
minerals = ['Mg2SiO4', 'MgSiO3']
components = ['MgO', 'SiO2']
mol_mass = {}
for c in components:
    mol_mass[c] = pt.formula(c).mass
for c in minerals:
    mol_mass[c] = pt.formula(c).mass

mol_mass
```

```python
2 * mol_mass['MgO'] / mol_mass['Mg2SiO4']
```

```python
mol_mass['SiO2'] / mol_mass['Mg2SiO4']
```

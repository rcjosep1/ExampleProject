# Weight % Calculation from Mol Fraction of Minerals

This notebook shows the first part of the workflow for converting mineral chemistry into oxide weight fractions. Here, we use Mg-endmember olivine and pyroxene as simple examples and calculate how much `MgO` and `SiO2` each mineral contributes by mass.

```python
python -c "import sys; print(sys.executable)"
```

## 1. Import the libraries

`numpy` is available for future numeric work, and `periodictable` is used to look up molar masses directly from chemical formulas.

```python
import numpy as np
import periodictable as pt
```

## 2. Define the mineral formulas and oxide breakdown

We treat the two minerals as idealized magnesium endmembers:

- Olivine: `Mg2SiO4`
- Pyroxene: `MgSiO3`

To express them as oxide components, we use:

- `Mg2SiO4 -> 2MgO + SiO2`
- `MgSiO3 -> MgO + SiO2`

These relationships tell us how many moles of each oxide are produced from one mole of mineral.

## 3. Quick mass check for one mineral

This cell verifies the molar mass of olivine (`Mg2SiO4`) before we build the full table of masses.

```python
pt.formula('Mg2SiO4').mass
```

## 4. Build a lookup table of molar masses

We collect molar masses for both the oxide components (`MgO`, `SiO2`) and the mineral formulas (`Mg2SiO4`, `MgSiO3`). This makes the later fraction calculations easier to read.

```python
minerals = ['Mg2SiO4', 'MgSiO3']
components = ['MgO', 'SiO2']

mol_mass = {}
for c in components:
    mol_mass[c] = pt.formula(c).mass

for c in minerals:
    mol_mass[c] = pt.formula(c).mass

mol_mass
```

## 5. Calculate the `MgO` weight fraction in olivine

One mole of `Mg2SiO4` contains the equivalent of **2 moles of `MgO`**. We therefore divide the mass of `2 x MgO` by the mass of one mole of olivine.

The result is a fraction, so multiply by 100 if you want wt%.

```python
2 * mol_mass['MgO'] / mol_mass['Mg2SiO4']
```

## 6. Calculate the `SiO2` weight fraction in olivine

One mole of `Mg2SiO4` contains the equivalent of **1 mole of `SiO2`**. So this fraction is the mass of `SiO2` divided by the mass of one mole of olivine.

```python
mol_mass['SiO2'] / mol_mass['Mg2SiO4']
```

## 7. Calculate the `MgO` weight fraction in pyroxene

One mole of `MgSiO3` contains the equivalent of **1 mole of `MgO`**. We therefore divide the mass of `MgO` by the mass of one mole of pyroxene.

As above, the result is a mass fraction and can be multiplied by 100 to report wt%.

```python
mol_mass['MgO'] / mol_mass['MgSiO3']
```

## 8. Calculate the `SiO2` weight fraction in pyroxene

One mole of `MgSiO3` also contains the equivalent of **1 mole of `SiO2`**. So this fraction is the mass of `SiO2` divided by the mass of one mole of pyroxene.

```python
mol_mass['SiO2'] / mol_mass['MgSiO3']
```

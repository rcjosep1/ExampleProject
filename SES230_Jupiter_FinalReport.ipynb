## Data sources and workflow

We assembled a multi-dataset geochemical compilation to compare basaltic compositions through time across tectonic settings. The core datasets were:

- **MORB**: mid-ocean ridge basalt data for divergent settings
- **OIB / intraplate volcanics**: hotspot-like and intraplate settings
- **ARC**: convergent-margin volcanic rocks
- **ARC_AGE**: a companion age dataset used to extend temporal coverage where chemistry and age were not recorded in the same file

### Main challenge
The source files were not directly comparable. Column names differed across datasets, and chemistry, tectonic setting, and age information were often split across different files. Some datasets contained only chemistry, some contained only age, and some contained both but with inconsistent naming conventions.

### Data cleaning strategy
We standardized each dataset separately and then concatenated the cleaned tables into one unified dataframe. Ages were harmonized using a consistent rule:

- if a direct age value was available, use it;
- if only minimum and maximum ages were available, use their midpoint;
- if only one bound was available, use that bound;
- if no usable age existed, drop the sample from the time-series plot.

### Output products
From the cleaned dataset we generated:
- a merged CSV file for reuse,
- binned age summaries,
- median geochemical trends through time,
- interquartile shading around the median lines,
- and publication-style figures for SiO2, MgO, and TiO2.

### Interpretation caveat
The datasets do not have equal temporal coverage across settings. MORB is inherently skewed toward young ages, while intraplate and arc datasets provide broader but still uneven age distributions. This asymmetry reflects geological preservation and source dataset structure rather than a coding error.

## Qin dataset: targeted citation filtering

The Qin compilation contains a very large number of samples and citations. Rather than using it wholesale, we treat it as a curated source of additional geochemical observations. The preferred workflow is to select only citations and tectonic settings that clearly improve coverage of the project's target settings and time intervals.

### Qin selection principle
We favor Qin samples that:
- belong to a tectonic setting relevant to the project,
- come from citations that clearly map to a volcanic or mantle process of interest,
- and add age or regional coverage not already represented in the main datasets.

### Why not use all Qin rows?
The Qin file is too broad to include without filtering. It contains many repeated analyses per sample and a large diversity of source papers, so a manual citation-level selection is necessary to keep the final dataset interpretable and scientifically defensible.



## Data Integration: Additional Intraplate Datasets

### Overview

To improve temporal coverage of intraplate (hotspot) systems, two targeted datasets were incorporated:

- **Tornare (2016)** — Canary Islands volcanism  
- **McCoy-West (2010)** — Lookout Volcanics, New Zealand  

These datasets were selected because they:
- Provide whole-rock geochemistry (SiO₂, MgO, TiO₂)
- Represent well-constrained volcanic systems
- Include defined age ranges suitable for time-series analysis

---

## Tornare Dataset

**Source:** Appendix 1 (whole-rock lava compositions)

**Geological Context:**
- Setting: Ocean island / intraplate
- Age range: ~14–25 Ma (Miocene)

**Processing:**
- Extracted SiO₂, TiO₂, MgO from matrix-style spreadsheet
- Assigned:
  - Age_Ma = 19.5 (midpoint)
  - Setting = HOTSPOT

**Assumptions:**
- Samples represent a single volcanic system
- Midpoint age approximates eruption timing

**Limitations:**
- No per-sample ages
- Includes some evolved compositions

---

## McCoy-West Dataset (Lookout Volcanics)

**Source:** Appendix 4 (whole-rock XRF/ICP-MS data)

**Geological Context:**
- Setting: Continental intraplate volcanism
- Age range: ~82–100 Ma (mid-Cretaceous)

**Processing:**
- Extracted SiO₂, TiO₂, MgO from matrix-style spreadsheet
- Removed rows without usable chemistry
- Filtered cumulative samples:
  - MgO < 10 wt%
- Assigned:
  - Age_Ma = 91 (midpoint)
  - Setting = HOTSPOT

**Assumptions:**
- Samples represent erupted basaltic compositions
- Midpoint age represents the volcanic sequence

**Limitations:**
- No per-sample ages
- Some duplicate sample variants retained

## Conceptual Approach

This project uses a **population-level approach**:

> Samples are treated as representative of volcanic systems within defined age intervals, rather than individually dated points.

The inclusion of Tornare and McCoy datasets:
- Extends hotspot coverage to both ~20 Ma and ~90 Ma
- Provides a more balanced comparison across tectonic settings
- Maintains a transparent and reproducible methodology
